Amplitude Shift Keying (ASK)

```python
import numpy as np
from scipy import signal

import matplotlib.pyplot as plt

# Parámetros de la señal
f_carrier = 10  # Frecuencia de la portadora en Hz
f_bit = 2       # Frecuencia del bit en Hz
bit_rate = 1 / f_bit  # Velocidad de transmisión en segundos por bit
num_sample_per_bit = 100
amplitude_1 = 0.5   # Amplitud para bit 1
amplitude_0 = 0.1   # Amplitud para bit 0

# Mensaje a modular (en binario)
message = '1011001'

# Convertir el mensaje a una lista de bits
binary_data = [int(bit) for bit in message]

# Duración de cada bit
t = np.linspace(0, bit_rate, num_sample_per_bit )

# Señal modulada ASK
modulated_signal = np.array([])
modulation_signal = np.array([])
for bit in binary_data:
    if bit == 0:
        modulated_signal = np.append(modulated_signal, amplitude_0 * np.sin(2 * np.pi * f_carrier * t))
        modulation_signal = np.append(modulation_signal, [0 for i in t])        
    else:
        modulated_signal = np.append(modulated_signal, amplitude_1 * np.sin(2 * np.pi * f_carrier * t))
        modulation_signal = np.append(modulation_signal,  [1 for i in t])
        
        
# Tiempo total de la señal
total_time = len(binary_data) * bit_rate

# Eje de tiempo para la señal modulada
time = np.linspace(0, total_time, len(modulated_signal))



# Graficar la señal modulada ASK

plt.figure(1)

plt.subplot(211)
plt.title('Modulación ASK')
plt.plot(time, modulation_signal)
plt.subplot(212)
plt.plot(time, modulated_signal)

plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
f_carrier   = 10e6  # Frecuencia de la portadora en Hz
f_bit       = 100       # Frecuencia del bit en Hz
bit_rate    = 1 / f_bit  # Velocidad de transmisión en segundos por bit
amplitude_1 = 0.5   # Amplitud para bit 1
amplitude_0 = 0.1   # Amplitud para bit 0

# Muestreo del DAC
sample_rate = 125e6  # Muestras por segundo (125Msps)

# Mensaje a modular (en binario)
message = '101100110011'

# Convertir el mensaje a una lista de bits
binary_data = [int(bit) for bit in message]

# Duración de cada bit
t_bit = np.arange(0, bit_rate, 1/sample_rate)

# Señal modulada ASK
modulated_signal = np.array([])
for bit in binary_data:
    if bit == 0:
        modulated_signal = np.append(modulated_signal, amplitude_0 * np.sin(2 * np.pi * f_carrier * t_bit))
    else:
        modulated_signal = np.append(modulated_signal, amplitude_1 * np.sin(2 * np.pi * f_carrier * t_bit))

# Tiempo total de la señal
total_time = len(binary_data) * bit_rate

# Eje de tiempo para la señal modulada
time = np.linspace(0, total_time, len(modulated_signal))

# Graficar la señal modulada ASK
plt.plot(time, modulated_signal)
plt.title('Modulación ASK')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```
