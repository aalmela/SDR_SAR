Phase Shift Keying (PSK)

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
f_carrier = 10  # Frecuencia de la portadora en Hz
f_bit     = 5       # Frecuencia del bit en Hz
bit_rate  = 1 / f_bit  # Velocidad de transmisión en segundos por bit
amplitude = 1     # Amplitud de la señal portadora

# Mensaje a modular (en binario)
message = '101100'

# Convertir el mensaje a una lista de bits
binary_data = [int(bit) for bit in message]

# Duración de cada bit
t = np.linspace(0, bit_rate, 1000)

# Señal modulada ASK
modulated_signal = np.array([])
for bit in binary_data:
    if bit == 0:
        modulated_signal = np.append(modulated_signal, amplitude * np.sin(2 * np.pi * f_carrier * t))
    else:
        modulated_signal = np.append(modulated_signal, -amplitude * np.sin(2 * np.pi * f_carrier * t))

# Tiempo total de la señal
total_time = len(binary_data) * bit_rate

# Eje de tiempo para la señal modulada
time = np.linspace(0, total_time, len(modulated_signal))

# Graficar la señal modulada ASK
plt.plot(time, modulated_signal)
plt.title('Modulación PSK')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
f_carrier   = 10e6  # Frecuencia de la portadora en Hz
f_bit       = 100       # Frecuencia del bit en Hz
bit_rate    = 1 / f_bit  # Velocidad de transmisión en segundos por bit
amplitude   = 0.5   # Amplitud para bit

# Muestreo del DAC
sample_rate = 125e6  # Muestras por segundo (125Msps)

# Mensaje a modular (en binario)
message = '101100110011'

# Convertir el mensaje a una lista de bits
binary_data = [int(bit) for bit in message]

# Duración de cada bit
t_bit = np.arange(0, bit_rate, 1/sample_rate)

# Señal modulada PSK
modulated_signal = np.array([])
for bit in binary_data:
    if bit == 0:
        modulated_signal = np.append(modulated_signal,  amplitude * np.sin(2 * np.pi * f_carrier * t_bit))
    else:
        modulated_signal = np.append(modulated_signal, -amplitude * np.sin(2 * np.pi * f_carrier * t_bit))

# Tiempo total de la señal
total_time = len(binary_data) * bit_rate

# Eje de tiempo para la señal modulada
time = np.linspace(0, total_time, len(modulated_signal))

# Graficar la señal modulada ASK
plt.plot(time, modulated_signal)
plt.title('Modulación PSK')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```
