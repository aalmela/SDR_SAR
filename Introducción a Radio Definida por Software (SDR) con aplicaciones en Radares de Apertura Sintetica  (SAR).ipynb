#  Introducción a la Radio Definida por Software (SDR) con aplicaciones en Radares de Apertura Sintética  (SAR)

## Resumen:

Este curso proporciona una introducción a los sistemas de Radio Definida por Software (SDR), ejemplos de su implementación y aplicaciones en Radares de Apertura Sintética (SAR).
SDR combina elementos de sistemas de radio y procesamiento de señales digitales para crear sistemas con aplicaciones en radiocomunicaciones, radares o sistemas de multiplexado en frecuencia, entre otros.  
SAR es una técnica de tele-observación que emplea señales de microondas, sistemas de antenas con apertura sintética, el procesamiento de señales y el procesamiento de información de imágenes. Este curso brinda una introducción a los esquemas básicos de hardware para SDR para el procesamiento de señales de radio y  aplicaciones en SAR. 

El curso comienza con la introducción al SDR y teoría de las comunicaciones. Luego el curso cubre programación en el entorno Python a través de ejemplos de SDR. 
Con el fin de conocer las limitaciones del diseño, se abordarán los conceptos de ruido,  distorsión y no linealidad. 
Por último, se realizará una revisión de las técnicas de digitalización y generación. 
Durante el curso se realizarán simulaciones con lenguajes como Python los cuales se implementaran empleando placas de evaluación SDR.  Los ejemplos se utilizan para demostrar las funciones básicas de procesamiento de señales SDR. 

## Requisitos previos por temas:
   1. Comprensión de la propagación y características básicas de las ondas electromagnéticas.
   2. Comprensión de la teoría básica de las comunicaciones.
   3. Comprensión de la teoría básica del procesamiento de señales digitales.

## Los estudiantes que completen este curso serán capaces de:
   1. Identificar los fundamentos del enlace de comunicación y ser capaces de discutir la asignación de recursos y tecnologías de radio.
   2. Comprender los sistemas que emplean sistema de radio definido por software y las limitaciones en el diseño de un sistema de radio definido por software.
   3. Comprender los fundamentos del diseño de sistemas para el diseño de un sistema de radiocomunicaciones definido por software.
   4. Comprender cómo se utilizan las tecnologías analógicas y digitales para las radios definidas por software y las topologías y aplicaciones.
